### Function 함수

return문
- return문이 있는 함수는 그 결과값을 생성하고, 그 시점에서 함수 실행 중지됨
- return문이 없는 함수는 여전히 실행되지만 None을 반환함
- return문을 사용할지 여부는 함수가 코드의 다른 곳에서 사용할 수 있는 생성하기를 원하는지 여부에 달려있음
- 결과를 반환할 필요가 없다면 반환문이 없어도 괜찮음

```python
# 1. string argument 문자열 인자

## a
def add_txt(t1, t2):   # string을 이용했을 때, return값이 없는 함수
  print(t1+t2)

## b
def add_txt(t1, t2):   # string을 이용했을 때, return값이 있는 함수
  result= t1+t2
  return result
```

```python
# 1. b
def add_txt(t1, t2):
  result= t1+t2
  return result

# 두 문자열로 인수 정
text1= "Hello, "
text2= "World!"

# 정의된 인수를 사용하여 add_txt 함수 호출
combined_text= add_txt(text1, text2)

print(combined_text)
```

```python
# 2 number argument 숫자 인자

# a
def add_number(n1, n2):
  print(n1+n2)   # numbers를 이용했을 때, return값이 없는 함수

# b
def add_number(n1,n2):   # numbers를 이용했을 때, return값이 있는 함수
  result= n1+n2
  return result
```

### Place holders (자리 표시자)

- %d 및 %s: 문자열 형식에 사용되는 자리 표시자
- 숫자나 문자열과 같이 특정 값을 삽임할 위치를 나타내기 위해 문자열에 사용됨.
  - %d: 정수(10진수)를 포맷하기 위한 자리 표시자. 문자열에서 %d를 사용하면 정수값으로 바뀜
  - %s: 문자열을 포맷하기 위한 자리 표시자, 문자열에 %s를 사용하면 문자열값으로 바뀜

```python
name= "Alice"
score= 95
age= 18

# %d 자리에 score 변수값을 문자열에 삽입
test_result= "You scored %d on the test." %score
print(test_result)

# %s와 %d를 사용하여 name과 age 변수값을 문자열에 삽입
greeting= "Hello, %s! You turn to finally %d years old finally." %(name, age)
print(greeting)

print("You scored %d on the test." %score)
print("Hello, %s! You turn to finally %d years old finally." %(name, age))
```

```python
# 2b
def add(a,b):
  result= a+b
  return result

num1= 5
num2= 3

sum_result= add(num1, num2)

print("The sum of %d and %d is %d" %(num1, num2, sum_result))
```

```python
# 2b_2

def add_and_print(a, b):
    result = a + b
    print("The sum of %d and %d is %d" % (a, b, result))
    return result

# a=3, b=4
sum_result = add_and_print(3, 4)
```

- 키워드 인자: 함수 호출할 때 괄호 안에 등호를 사용하여 표시 (ex: add_txt1(t1= 'My freind', t2= 'Python')
- 기본 인자: 함수 호출할 때 t2위치에 실제값을 전달하지 않아도 'Python'이 기본값으로 대입됨

```python
# 위의 함수 정의 및 호출 이해 완료

# add_txt2: t1, t2를 :로 연결 후 출력하는 함수
def add_txt1(t1, t2= 'Python'):
  print(t1 + ':' + t2)

# add_txt1 함수 호출
##기본값 t2='Python'을 사용하며 호출
add_txt1('Best')

## 인자 이름을 명시하며(키워드인자 사용) 호출
add_txt1(t1= 'My freind', t2= 'Python')
add_txt1(t1= 'My buddy')
```

- 가변 인자(*args)는 인자의 개수가 명확하지 않을 때 사용
  - 가변 길이 인수(즉, varargs)를 사용하면 함수에 서로 다른 수의 인수 전달 가능
  - *args매개 변수는 함수에 전달된 모든 위치 인수를 수집하고 함수 내에서 tuple로 액세스할 수 있도록 함.

```python
def func_uncertain(*args):
  print(args)

# *args에 아무값도 전달하지 않고 함수를 호출하면 빈 tuple로 결과가 출력됨
func_uncertain()

# 위치 인수의 개수가 다른 함수 호출
func_uncertain(1,2,3)
func_uncertain('apple', 'banana', 'cherry', 'date')
func_uncertain(True)
```

- 키워드 가변 인자 (**kwargs)는 차후 어떤 키워드 인자가 필요한지 미확정 되었을 때 사용
  - 키워드 인수 또는 명명된 인수는 매개 변수 이름을 값과 함께 지정하는 방식으로 함수에 인수를 전달하는 방법
  - 키워드 인수를 사용하여 값을 제공하는 매개 변수의 이름을 명시적으로 지정하여 함수 호출을 더 명확하고 자체 문서화 가능

```python
# 함수를 정의할 때 TWO asterisks인 것 주의
# 인자는 string이 아니기 때문에 ''를 동반 사용 X

def func_kwa(width, height, **kwargs):
  print(kwargs)

# *kwargs에 아무값도 전달하지 않으면서 함수를 호출하면 empty dictionary로 결과 출력
func_kwa(10, 20)

# kwargs(키워드인자) 두개 추가
func_kwa(10, 20, depth=30, color='blue')
```
