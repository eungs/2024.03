### Assigning Dictionaries to variables

```python
k = {'Mercury':'수성', 'Venus':'금성', 'Earth':'지구', 'Mars':'화성', 'Jupiter':'목성', 'Saturn':'토성', 'Uranus':'명왕성', 'Neptune':'해왕성'}

m = {'Mercury 수성':1, 'Venus 금성':2, 'Earth 지구':3, 'Mars 화성':4, 'Jupiter 목성':5, 'Saturn 토성':6, 'Uranus 명왕성':7, 'Neptune 해왕성':8}

print (k)
print (m)
```

```python
dict = {'a':1, 'b':2, 'c':3}
print (dict)
print(dict['a'])   # dict 중 a만 출력
```

```python
dict = {'a':1, 'b':2, 'c':3}

print(dict)

dict['a']=7   # dict 중 a값을 7로 변경
print (dict)

dict_data = {'c':10, 'b':20, 'a':30}
print(dict_data)
dict_data['a'] = 300   # dict_data 중 a값을 300으로 변경
print(dict_data)
```
