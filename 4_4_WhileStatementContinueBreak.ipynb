### While statement Loop
- 특정 조건이 사실인 코드를 반복적으로 실행
- 지정된 조건이 거짓이 될 때까지 반복적으로 작업을 수행할 루프 생성

While Statement 1

```python
counter= 1

while counter <= 5:
  print(counter)
  counter += 1   # 1부터 5까지 출력

print("Loop finished!")   # 5까지 출력되면 print구문 출력
```

```python
x= 1
total= 0

while 1:
  total= total + x   # 기존에 설정한 'total=0'에 x인 1을 더한 값을 total로 다시 지정

  if total > 100000:  -
    print(x)
    print(total)
    break
```
