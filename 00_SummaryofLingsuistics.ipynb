## Human Language Speech: Communication Chain

### The speech communication chain
1. Think of what you want to communicate.
  - 소통하고 싶은 것 생각하기
2. Pick our words to express the idea.
  - 아이디어를 표현할 단어 고르기
3. Put there words together in a certain order following reuls.
  - 규칙에 따라 일정한 순서대로 단어들 정리하기
4. Figure out how to pronounce these orders.
  - 이 단어들이 어떻게 발음하는지 알아보기
5. Send these pronounciations to your vocal anatomy.
  - 그 발음들을 음성 해부학으로 보내기
6. Speak: Send the sounds through the air.
  - 말하기: 공기를 통해 소리 내보내기
7. Perceive: Listener hears the sounds
  - 인지: 청취자가 소리 듣기
8. Decode: Listener interprets sounds as language.
  - 디코딩: 듣는 사람은 소리를 언어로 해석
9. Connect: Listener receives communicated idea.
  - 연결: 청취자가 전달된 아이디어 수신

## Morphology (형태학)
You know grammatical categories of words irrespective of whether you are aware of them or not as a speaker of English language.
  - 영어권의 화자로서 인지 여부와 상관없이 단어의 문법적 범주를 알고 있음.

- Morpheme (형태소): 의미의 최소 단위



- Root (어근):
  - 어근은 단독으로 서 있을 수도 있고 접사와 결합되어 단어를 형성 가능
  - Free Morphemes (자립형태소)
    - Cook cooker; paint, painter; believe, believable
  - 뿌리가 단독으로 서 있지 않을 때 그것은 속박된 뿌리라고 불림.
    - root may not stand alone: Bound Morphemes (외종형태소)
    - -ceive (receive, conceive, perceive, deceive)
    - -mit (remit, permit, commit, submit, transmit, admit)
    - -cern (concern, discern)
    - cran- (cranberry)


- Stem (어간):
  - 접사와 결합된 형태소 (접사)
    - cook-er, paint-er, believ-able, system-atic


- Affix (접사):
  - 어떤 형태소는 그 자체로 단어를 구성하는 것이 아니라 항상 병동의 일부를 구성함.
  - 접두사의 위치에 따라:
    - Prefix (접두사): 다른 형태소보다 먼저 발생
      - e.g.s, prefer, impossible, conclude, etc.
    - Suffix (접미사): 다른 형태소 뒤에 붙음
      - e.g., solves, solved, solution, etc.
    - Infix (삽입사): 다른 형태소에 삽임
      - e.g.s, abso-bloomin-lutely in BE

### 자연어 처리를 위한 예제 단위

- Compounds (화합물)
  - 화합물에서 가장 오른쪽에 있는 단어: 화합물의 머리
  - 머리: 단어나 구의 넓은 의미와 문법 범주를 결정하는 부분
    - 'headstrong (N + A = A)
    - 'pickpocket (V + N = N)

  - 전치사와 함께 형성된 화합물은 화합물의 전치사가 아닌 부분의 범주에 속함.
    - over'take(Prep + V = V)
    - 'sundown(N + Prep = N)

  - 전치사가 새로운 구성원을 쉽게 인정하지 않는 폐쇄적인 계급 범주를 형성한다는 추가적인 증거
