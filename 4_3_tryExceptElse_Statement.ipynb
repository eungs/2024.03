###try ~ except ~ else/finally

1. 프로그램 작성 후 실행되는 동안 오류 발생하면 프로그램 진행 X, 이 상황을 예외(exception)상황이라 함.


2. try: ~ except:
 - Exception이 있더라도 중단없이 예외에 대하여 적절히 처리하여 프로그램을 진행시킬 수 있도록 하는 구문


3. try: ~ except: ~ else:
 - 어떤 로직을 수행했을 때 오류 상황이 아닌 경우에만 어떤 작업을 수행하는 코드를 작성할 때
 - Exception이 발생하면 else 부분은 수행되지 않음


4. try: ~ except: ~ finally:
 - 오류 발생과 상관없이 어떤 코드를 무조건 실행시킬 때

```python
# '2. try: except:' 예시

try:
  print('How are you?')
  print(param)   # param변수가 정의 X, Exception(예외)발생

except:
  print('Exception occurred')
```

```python
# '3. try: ~ except: ~ else:' 예시 1 (else부분 실행 X 경우)

try:
  print('How are you?')
  print(param)

except:
  print('Exception occurred')

else:
  print('Exception did not occur.')
```

```python
# '3. try: ~ except: ~ else:' 예시 1 (except부분 실행 X 경우)

try:
  print('How are you?')

except:
  print('Exception occurred')

else:
  print('Exception did not occur.')
```

```python
# '4. try: ~ exception: ~ finally: exception'

try:
  print ('How are you?')
#  print (param)

except:
  print ('Exception occurred.')

finally:
  print ('This code should be running.')
```
