## **Place holders - 자리표시자**

- %d: 숫자 자리표시자
- %s: 문자열 자리표시자

```python
# 3. Example of place holders such as %d and %s

name= "Alice"
score= 95
age= 18

# Using %s and %d for string formatting
test_result= "You scored %d on the test." %score
print(test_result)
```

```python
# Example of place holders such as %d and %s

name= "Alice"  # 문자
score= 95  # 숫자
age= 18  # 숫자

# Using %s and %d for string formatting
test_result= "You scored %d on the test." %score  # 숫자
print(test_result)

greeting = "Hello, %s! You turn to finally %d years old finally." %(name, age)  # 문자, 숫자
print(greeting)
print('\n')

# Alternatively
print("You scored %d on the test." %score)  # 숫자
print("Hello, %s! You turn to finally %d years old finally." %(name, age))  # 문자, 숫자
```

```python
# 2.b

def add(a, b):
  result= a+b
  return result

num1= 5
num2= 3

sum_result= add(num1, num2)

print("The sum of %d and %d is %d" %(num1, num2, sum_result))
```

```python
# 2.b'

def add_and_print(a, b):
  result= a+b
  print("The sum of %d and %d is %d" %(a, b, result))
  return result

sum_result= add_and_print(3, 4)
```

- 키워드 인자: 함수 호출 때 괄호안에 등호를 사용하여 표시
  - add_txt1(t1= 'My friend', t2= 'Python)
- 기본 인자: 함수 호출 때 t2 위치에 실제값을 전달하지 않아도 'Python'이 기본값으로 대입
  - t2= 'Python'

```python
# 함수 호풀

def add_txt1(t1, t2= 'Python'):  # add_txt1은 t1과 t2를 ':'로 연결 후 추력
  print(t1 + ':' + t2)  # 보지마삼

add_txt1('Best')  # 기본값 t2='Python' 사용하여 add_txt1 함수 호출
add_txt1(t1= 'My friend', t2= 'Python')  # 인자 이름을 명시하며(키워드 인자 사용) add_txt1 함수 호출
add_txt1(t1= 'My Buddy')
```

### **가변 인자(*args)**
- 인자의 개수가 명확하지 않을 때 사용

```python
def func_uncertain(*args):
  print(args)

func_uncertain()  # *args에 아무값도 전달하지 않고 함수를 호출하면 빈 튜플로 결과 출력

func_uncertain(1, 2, 3)
func_uncertain('apple', 'banana', 'cherry', 'date')
# func_uncertain(True)  # 결과값에 (True,)가 나오는 이유 모름
```

### **키워드 가변 인자(*kwargs)**
- 차후 어떤 키워드 인자가 필요한지 미확정 되었을 떄 사용

```python
def func_kwa(width, height, **kwargs):  # 인자는 string이 아니기 때문에 ''룰 동반 사용 X
  print(kwargs)

func_kwa(10, 20)  # **kwargs에 아무것도 전달하지 않으면서 호출하면 빈값 출력
func_kwa(10, 20, depth=30, color='blue')  # 키워드인자 두개 추가
```
