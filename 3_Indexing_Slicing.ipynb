```python
list_data = ['apple', 'banana', 'grapes', 'mango']
list_data[0]="shine musket"   # 0번째인 'apple'을 'shine musket'으로 변경

print (list_data)
```

```python
str_data = 'Time is money!!'
print(str_data[1:5])   # 1번째부터 4번째는 'ime  '
print(str_data[:7])   # 0번째부터 6번째는 'Time is'
print(str_data[9:])   # 9번째부터 마지막까지는 'oney!!'
print(str_data[:-3])   # 처음부터 뒤에서 2번째(0,1,2) 제외까지는 'Time is mone'
print(str_data[-3:])   # 뒤에서 3번째부터 마지막까지는 'y!!'
print(str_data[:])   # 처음부터 끝까지
print(str_data[::2]) # 처음부터 끝까지 스텝을 2로 해서 슬라이싱 하기.

print(str_data[1:3])
```
