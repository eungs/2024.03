class의 맥락에서 "instant object" 또는 "instance object"를 언급할때 해당 클래스에서 생성된 특정 개체임

클래스의 개체 또는 인스턴스는 고유한 데이터를 가지며 클래스 메서드에 의해 정의된 작업 수행 가능

- 객체(object) or 인스턴스(instance)
  - 클래스로 만든 객체를 '인스턴스' 라고 함
  

```python
# 'FourCal'이라는 이름의 class 생성
class FourCal:

  # 클래스 내에서 method(클래스 내에서 정의된 함) 'setdata' 정의
  def setdata(self, first, second):

    # 'first' 매개 변수를 인스턴스 변수 first에 할당
    ## 객체에 생성되는 개체만의 변수를 '객체변수' 또는 '속성'
    ## a 객체에 객체변수 first가 생성되고 4라는 값 저
    self.first= first

    # 'second' 매개 변수를 인스턴스 변수 'second'에 할
    ## a 객체에 객체변수 second가 생성되고 2라는 값 저장
    self.second= second
```

위와 같이 '클래스명.메서드' 형태로 호출할 때: 객체 a를 첫번째 매개변수 self에 꼭 전달해야함


    a = FourCal()
    a.setdata(4, 2)
'객체.메서드' 형태로 호출할 때: slef를 반드시 생략해서 호출해야함.

```python
# 클래스 만들기: 확장1 -> addition

class FourCal:
  def setdata(self, first, second):
    self.first = first
    self.second = second

  def add(self):
    result = self.first + self.second
    return result
```

```python
# 객체 a, 객체변수 first, 객체변수 second 할당, add 연산

a = FourCal()
a.setdata(10,20)
a.add()
```

```python
# 클래스 만들기: 확장2 -> subtraction, multiplicaton, division

class FourCal:
  def setdata(self, first, second):
    self.first= first
    self.second= second

  def add(self):
     result= self.first + self.second
     return result

  def sub(self):
    result= self.first - self.second
    return result

  def mul(self):
    result= self.first * self.second
    return result

  def div(self):
    result= self.first / self.second
    return result

a= FourCal()
b= FourCal()

a.setdata(10, 20)
b.setdata(300, 400)
```

```python
a.add()   # 10 + 20
```

```python
print(b.add())   # 300 + 400
print(a.sub())   # 10 - 20
print(b.sub())   # 300 - 400
print(a.mul())   # 10 * 20
print(b.mul())   # 300 * 400
print(a.div())   # 10 / 20
print(b.div())   # 300 / 400
```

#### Class Inheritance
- 기존 클래스가 라이브러리 형태로 제공되거나 수정이 허용되지 않는 상황이라면 상속을 사용
- 아래 예시는 제곱(power) 메소드

```python
class MoreFourCal(FourCal):
  def pow(self):
    result= self.first ** self.second
    return result

a= MoreFourCal()
```

```python
a.setdata(10, 20)
a.pow()   # 10의 20제곱
```

```python
class Greetings:   # 독립된 공간을 Class 라고 함
  def sayHello(self):   # Class Method
    print('How are you?')

# Class Method에서 두 번째 인자로 사용되는 name은 Instant Member
  def sayBye(self,name):
    print('%s! see you!' %name)

obj = Greetings()
obj.sayHello()
obj.sayBye('Emma')
```

```python
# 글로벌변수 이름 결과 및 값 0 할당
result= 0

# add함수: 하나의 매개변수인 num을 사용
def add(num):

  # 덧셈 함수 내에서 전역 변수 결과를 사용할 것임을 선언
  # 이를 통해 함수 내에서 글로벌 결과 변수 수정 가능
  global result

  # result에 num 더하기
  # num 매개 변수의 값을 전역 결과 변수에 추가
  # 여러 함수 호출에 대한 값을 누적하여 실행 중인 값 반환
  result += num
  return result

print(add(3))
print(add(4))
print(add(5))
```

```python
class Calculator:

  # 초기화 방법을 사용하여 self.result의 초기값을 0으로 설정
  def initialize(self):
    self.result = 0

  def add(self, num):
    self.result += num
    return self.result

  def sub(self, num):
    self.result -= num
    return self.result

# Calculator class의 인스턴스 생성
calc= Calculator()

# calculator 결과 초기화
calc.initialize()

# 덧셈과 뺄셈 수행
result_add = calc.add(1)
result_sub = calc.sub(1)

print(result_add)
print(result_sub)
```
