### For ~: If ~: CONTINUE... ELSE: BREAK

```python
scope= [1,2,3,4,5]

for i in scope:
  print(i)
  if i < 3:   # i가 3보다 작을 경우 계속 진행해서 출력
    continue   # continue 없어도 코드 작성 가능

  else:   # i가 3보다 크거나 같을 경우 출력 중단
    break
```

### FOR ~: ELSE ~:

```python
scope= [1,2,3,4,5]

for x in scope:
  print(x)
  break

else:
  print('Perfect')
```
